# **Contexto del proyecto**

El Programa de Habitabilidad Rural (PHR en adelante), es un programa del Ministerio de Vivienda y Urbanismo que tiene por objetivo mejorar la calidad de vida de las familias que residen en áreas rurales, garantizando que sus viviendas cumplan con los estándares de habitabilidad establecidos por el Ministerio.

Este cuenta con tres componentes 1) Construcción de Vivienda Nueva, con las tipologías de proyectos Construcción en Sitio del Residente o Construcción de Conjunto Habitacional; 2) Mejoramiento y/o Ampliación de Vivienda Existente; 3) Mejoramiento de Entorno y Equipamiento Comunitario.

Los datos asociados a la ejecución del PHR son consolidados por parte de MINVU en un solo dataset denominado "Base PHR", que fue pedida a la institución vía solicitud de información pública en el marco de la Ley de Transparencia.

A partir de la observación inicial de los datos contenidos en el dataset se concluyó que los principales desafíos para la institución se encuentran en los tiempos de ejecución de los proyectos, razón por la cual, nuestra motivación será generar una optimización de los mismos.


El proyecto tiene como finalidad abordar las siguientes preguntas planteadas en el hito:

* ¿Se puede segmentar la información para llevar a cabo un análisis más detallado y descriptivo de cada componente del programa?
* ¿Dónde se localizan los cuellos de botella o los puntos críticos en términos de ejecución territorial?
* ¿Existen territorios o comunas particularmente desfavorecidos en la ejecución de proyectos, o, por el contrario, se han evaluado positivamente?
* ¿Es posible realizar predicciones sobre la cantidad de obras que se iniciarán y su tiempo de ejecución?

En resumen, este proyecto tiene como objetivo aprovechar la información recopilada para mejorar la planificación y ejecución del Programa de Habitabilidad Rural, garantizando que las familias rurales tengan acceso a viviendas que cumplan con los estándares de habitabilidad. Las preguntas planteadas buscan optimizar la toma de decisiones, aumentar la eficiencia en la ejecución de proyectos y identificar áreas que requieran atención especial en términos de desarrollo rural.

###Limpieza y corrección de fechas

Dado que los datos en su mayoria son ingresados manualmente, las fechas que contiene el archivo Excel presentan diversos formatos (como aaaa-mm-dd, mm/dd/yyyy, mm/dd/yy, entre otros).

También se identificaron casos en los que, debido a la falta de columnas y/o la imprevisibilidad de nuevas etapas que debían ser registradas, estas se agregaron a la tabla con un separador "+".

En los casos necesarios, se trabajaron las fechas para ajustarlas al formato datetime.

Las variables afectadas por esta situacion son las siguientes:

- Fecha Publicación DO
- Fecha RE Incr. Obra
- Fecha RE Complem.
- Fecha RE Aprobación Proyecto DITEC
- Fecha Término de Obra DITEC

```python
df['Fecha Publicación DO']=df.loc[:,'Fecha Publicación DO'].apply(pd.to_datetime,dayfirst=True)
fechas = df.loc[:,['Fecha RE Incr. Obra','Fecha RE Complem.','Fecha RE Aprobación Proyecto DITEC','Fecha Término de Obra  DITEC']]
mask = fechas['Fecha RE Incr. Obra'].str.find("+")>0
fechas[['fecha_incr_1','fecha_incr_2']] = fechas.loc[:,'Fecha RE Incr. Obra'].str.split("+",expand=True,).apply(lambda x: pd.to_datetime(x,dayfirst=True))
f_incr_aux = fechas.loc[~mask,'Fecha RE Incr. Obra'].apply(lambda x: pd.to_datetime(x,dayfirst=True))
fechas = fechas.join(f_incr_aux,how='left',rsuffix='_')
fechas['fecha_incr_1']=np.where(mask,fechas['fecha_incr_1'],fechas['Fecha RE Incr. Obra_'])
mask = fechas['Fecha RE Complem.'].str.find("+")>0
fechas[['fecha_complem_1','fecha_complem_2','fecha_complem_3']] = fechas.loc[:,'Fecha RE Complem.'].str.split("+",expand=True).apply(lambda x: x.str.strip()).apply(lambda x: pd.to_datetime(x,dayfirst=True))
f_compl_aux = fechas.loc[~mask,'Fecha RE Complem.'].apply(lambda x: pd.to_datetime(x,dayfirst=True))
fechas = fechas.join(f_compl_aux,how='left',rsuffix="_")
fechas['fecha_complem_1'] = np.where(mask,fechas['fecha_complem_1'],fechas['Fecha RE Complem._'])
fechas['fecha_aprob_proyecto_DITEC'] = fechas['Fecha RE Aprobación Proyecto DITEC'].apply(lambda x: x if isinstance(x,dt.datetime) else np.nan)
fechas['fecha_term_DITEC']=fechas.loc[:,'Fecha Término de Obra  DITEC'].apply(lambda x: x if isinstance(x,dt.datetime) else np.nan)
fechas.head()
```

En algunos casos, las informaciones son incoherentes o erróneas, sea por problemas sistémicos, o por la extración de los datos del sistema origen.
